# 🏥 Interpretable ICU Anomaly Attribution Pipeline Tutorial

Welcome! This notebook walks you through using the end-to-end, clinically-interpretable multivariate physiological anomaly detection and attribution pipeline.

## 🧬 Architecture Overview
The pipeline consists of the following phases:
1. **Ingestion & Generation**: Simulating or loading ICU data with circadian rhythms and physiological coupling.
2. **Cleaning & Preprocessing**: Handling missing values, filtering measurement artifacts, and computing composite indices (e.g., Shock Index).
3. **Decomposition**: Splitting signals into Trend, Seasonal, and Residual components via STL.
4. **Detection**: Identifying anomalies in residuals using a hybrid model (Z-score + Isolation Forest + LSTM Autoencoder).
5. **Attribution**: Explaining the anomalies using SHAP to pinpoint the key driving features.
6. **Relationship Analysis**: Computing contemporaneous correlations, lead-lag cross-correlations, and Granger Causality.
7. **Clinical Interpretation**: Translating the numerical results into plain-English narratives, active alerts, and prioritized clinical recommendations.
